# Images as Grids of Pixels

# Import Resources

```
# import resources

import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt 
import cv2

%matplotlib inline 
```

Car Image https://github.com/sayaliKutwal/CVND_Exercises/blob/master/1_1_Image_Representation/images/waymo_car.jpg

# Read and display the image

```
# Read the image
image = mpimg.imread('waymo_car.jpg')

# Print out thr image dimension
print('Image Dimension:', image.shape)

# Change from color to grayscale 
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Not assigning cmap=gray, image appears in different form 
plt.imshow(gray_image, cmap='gray')
```

# Find specific grayscale pixel value

```
# Prints specific grayscale pixel values
# What is the pixel value at x = 400 and y = 3000

x = 400
y = 300

gray_image[y, x]
```

# Find the maximum and minimum grayscale pixel value

When x > 427 or y > 640, error reads: *index a is out of bounds for axis 0/1 with size b*

```
# Finds the maximum and minimum grayscale values in htis image
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max: ', max_val)
print('Min: ', min_val)
```
